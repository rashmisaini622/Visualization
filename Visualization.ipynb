### Distribution plot
    distplot - check the distribtion of the columns feature
    joinplot - used for univarient or bivarient analysis i.e it is used to analysis the realtionship between two variable in         visualization form. 
    pairplot - when our number of features are more then three or we have multiple features then we use pairplot which show         variables paired with all other variables.

### Distplot

```python
sns.distplot(df['rating'])
```

### Joinplot

```python
sns.jointplot(x='user_id',y='rating',data=df,kind='hex')
```

### Boxplot
It measure of how well distributed is the data in a data set. It divides the data set into three quartiles.It is   also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.

```python
import seaborn as sns
#sns.boxplot(x="Age", y="Sex", data=df_T)
ax = sns.boxplot(x="Age", y="Sex", data=df_T)
ax.set_title("boxplot")                                                                 # Set title
```

### How To Set The Figure Size in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(10,4))
# Create swarmplot
sns.swarmplot(x="Sex", y="Age", data=df_T, ax=ax)
```

### How To Rotate Label Text in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import pandas as pd
# Initialize the data
x = 10 ** np.arange(1, 10)
y = x * 2
data = pd.DataFrame(data={'x': x, 'y': y})
# Create an lmplot
grid = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={"s": 100})
# Rotate the labels on x-axis
grid.set_xticklabels(rotation=30)
```
